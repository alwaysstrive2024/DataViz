```python
"""
Solve the differential equation y'(t) = y(t) using scipy
"""

from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exp(y, t):
    """
    Input: tuple y, floats t from model function for exponential
    
    Output: Estimates of derivative of y
    """
    x, = y
    dydt = x
    return dydt
```

```python
def plot_exp(t, sol):
    """
    Input: 1D numpy array t, 1D numpy array sol
    
    Action: Plot each coordinate of the solution vs t
    """

    plt.plot(t, sol, 'b', label='y(t)')
    plt.title("Solution of y'(t) = y(t) with y(0) = 1")
    plt.legend(loc='best')
    plt.xlabel('t')
    plt.grid()
    plt.show()
```

```python
def solve_exp():
    """ Solve an example """
    
    y0 = (1)
    t = np.linspace(0, 1, 101)
    sol = odeint(exp, y0, t)
    print(sol)
    plot_exp(t, sol)
    
solve_exp()
```

# Pend_ode example

```python
def pend(y,t,b,c):
    theta,omega = y
    dtheta_dt = omega
    domega_dt = -b*omega - c*np.sin(theta)
    return [dtheta_dt,domega_dt]
```

```python
b,c = 0.25,5.0
y0 = [np.pi-0.1, 0.0]
t = np.linspace(0, 10, 101)
sol = odeint(pend, y0, t, args=(b,c))
```

```python
print(sol)
```

```python
theta_values = sol[:, 0]  # 角度随时间变化
omega_values = sol[:, 1]  # 角速度随时间变化
```

```python
fig, ax = plt.subplots()
ax.plot(t, theta_values, 'b', label='theta(t)')
ax.plot(t, omega_values, 'r', label='omega(t)')
ax.grid(True)
ax.set_xlabel('t')
plt.show()
```
