```python
"""
Simple plots in matplotlib as specified via the matplotlib API
https://matplotlib.org/api/index.html
"""
import matplotlib.pyplot as plt
```

## pyplot API

An example of a simple plot using the [pyplot API](https://matplotlib.org/api/index.html#the-pyplot-api).

```python
plt.plot([0, 1, 2, 1, 0])
plt.title("Plot in matplotlib using pyplot API")
plt.savefig("simple_example_pyplot.png")
#plt.show()
```

An example of creating two subplots using the [pyplot API](https://matplotlib.org/api/index.html#the-pyplot-api). 

```python
plt.subplot(1, 2, 1)
plt.plot([0, 1, 2, 1, 0])
plt.title("Subplot 1 via pyplot API")
plt.subplot(1, 2, 2)
plt.plot([2, 1, 0, 1, 2])
plt.title("Subplot 2 via pyplot API")
#plt.show()
```

An example of encapsulating the plotting code in a plotting function using the [pyplot API](https://matplotlib.org/api/index.html#the-pyplot-api). Note that the function returns the `matplotlib` figure corresponding to the plot.

```python
def plot_data_plt(data, title):
    """
    Input: List data of floats, string title
    
    Output: matplotlib Figure object for line plot of data
    """
    
    fig = plt.figure()
    plt.plot(data)
    plt.title(title)
    return fig
```

Testing function that calls `plot_data_plt()`. Note the call to `plt.show()` is unnecessary.

```python
def test_plot_data_plt():
    """ Test plot_data_plt() """
    
    my_data = [0, 1, 2, 1, 0]
    plot_data_plt(my_data, "Plot in matplotlib using pyplot API")
    
test_plot_data_plt()
```

## Object-oriented API

An example of a simple plot using the [object-oriented API](https://matplotlib.org/api/index.html#the-object-oriented-api). 

```python
fig, axs = plt.subplots()
axs.plot([0, 1, 2, 1, 0])
axs.set_title("Plot in matplotlib using object-oriented API")
fig.savefig("simple_example_oo.png")
plt.show()
```

An example of creating two subplots using the [object-oriented API](https://matplotlib.org/api/index.html#the-object-oriented-api). 

```python
fig, (ax0, ax1) = plt.subplots(1, 2)
ax0.plot([0, 1, 2, 1, 0])
ax0.set_title("Subplot 1")
ax1.plot([2, 1, 0, 1, 2])
ax1.set_title("Subplot 2")
plt.show()
```

An example of encapsulating the plotting code in a plotting function using the [object-oriented API](https://matplotlib.org/api/index.html#the-object-oriented-api). Again, the plotting function returns a `matplotlib` figure.

```python
def plot_data_obj(data, title):
    """
    Input: List data of floats, string title
    
    Output: matplotlib Figure object for line plot of data
    """
    
    fig, axs = plt.subplots()
    axs.plot(data)
    axs.set_title(title)
    return fig
```

```python
def test_plot_data_obj():
    """ Test plot_data() """
    
    my_data = [0, 1, 2, 1, 0]
    plot_data_obj(my_data, "Plot in matplotlib using object-oriented API")
test_plot_data_obj()
```
