```python
"""
Examples of raster graphics, colormaps, and colors in matplotlib. 
"""

import matplotlib.pyplot as plt
import numpy as np
```

## Creating a raster image from an array

Here is an example of defining a numpy array manually and plotting it as an image using `imshow()`. For small arrays, each element in the array is rendered as a square of pixels. matplotlib automatically converts the integers to colors via a [colormap](https://matplotlib.org/tutorials/colors/colormaps.html). The default colormap in matplotlib is `viridis`.

```python
a_list = [[0, 1, 2, 3], 
          [3, 4, 5, 6], 
          [6, 7, 8, 9]]

an_array = np.array(a_list)
print(an_array)
print(an_array.shape)

plt.imshow(an_array)
plt.savefig("an_array_image.PNG")
plt.show()
```

## Reading an image as an array

Images can be loaded into matplotlib as numpy arrays using the `imread()` method.  Here is an example of a digitial elevation map of the Grand Canyon represented as a TIFF image loaded `imread()` and plotted using `imshow()`. 

```python
grand_canyon = plt.imread("../data/gc_dem_processed.tiff")
print("Loaded array from the image file is\n", grand_canyon)
print("Shape of loaded array is", grand_canyon.shape)
plt.imshow(grand_canyon, origin="lower")
plt.show()
```

Note that the shape of the array is 2000 (number of rows) by 3000 (number of columns), expressing its shape in terms of height x width. However, if you look at the properties of the input image `gc_dem_processed.tiff` from the data directory and explore the dimensions, its size is 3000 x 2000 (in pixels), expressing the size in terms of width x height. **Note that the width and height are reversed here.**

Also, note that the origin of image plots is in the upper left corner of the plot while, for line plots, matplotlib places the origin in the lower left. The position of the origin in image plots can be controlled via the option `origin="lower"`.

## Specifying a custom colormap and adding a colorbar

A custom colormap can be specified using the `cmap` option for `imshow()` (a list of available colormaps is [here](https://matplotlib.org/tutorials/colors/colormaps.html)). The chosen colormap can be visualized via a colorbar that can be added using the `plt.colorbar()` method.

Here is a plot of the Grand Canyon elevation map using the `RdBu_r` colormap.  Ths map uses blue to indicate "low" elevations and red to indicate "high" elevations.  Note the usage of `_r` to reverse the `RdBu` colormap.

```python
plt.imshow(grand_canyon, cmap="RdBu_r")
plt.colorbar()
plt.show()
```

## Specifying colors

There are three main methods to specify [color](https://matplotlib.org/2.0.2/api/colors_api.html) available in matplotlib:
* Predefined matplotlib [color strings](https://matplotlib.org/2.0.2/api/colors_api.html#module-matplotlib.colors)
* [RGB specification](https://en.wikipedia.org/wiki/RGB_color_model) as tuple of three floats in range [0, 1]
* [HTML hex string](https://en.wikipedia.org/wiki/Web_colors#Hex_triplet) - don't forget the leading `#`

Here are examples of each used in a plot of three lines.

```python
plt.plot([0, 1, 2, 1, 0], c="r")
plt.plot([2, 1, 0, 1, 2], c=(0, 1, 0))
plt.plot([1, 1, 1, 1, 1], c="#0000ff")
plt.show()
```
