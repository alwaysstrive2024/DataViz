```python
#######################################################################################
# Part 1 - Compute and plot Julia sets


def invert_fun(complex_fun, complex_val):
    """
    Input: np.poly1d object complex_fun and complex complex_val
    
    Output: List of n possible complex solutions to the equation complex_val = complex_fun(inv)
    The roots of complex_poly are the n possible values for complex_fun^(-1)(complex_val)
    """
    
    # Form the polynomial equation complex_fun(z) - complex_val = 0
    poly_equation = complex_fun - complex_val
    
    # Extract coefficients of the polynomial equation
    coefficients = poly_equation.coeffs
    
    # Compute the roots of the polynomial
    roots = np.roots(coefficients)
    
    # Sort the roots lexicographically (real parts first, then imaginary parts)
    sorted_roots = np.sort(roots)
    
    # Convert to a list of complex numbers and return
    return sorted_roots.tolist()
```

```python
def julia_set(lmbd, z_0, num_returned, num_dropped, seed=None):
    """
    Input: complex lmbd, z_0, integers num_returned, num_dropped,
    optional integer seed
    
    Output: List of complex numbers (starting with z_0)
    generated by Julia iteration with given lmbd
    """
    
    if seed:
        random.seed(seed)

    # Remember to use random.choice() when selecting an inverse
    
    
    
    # Initialize the sequence with the starting point z_0
    sequence = [z_0]
    
    # Calculate the total number of steps needed (including the initial z_0)
    total_steps = num_dropped + num_returned - 1
    
    for _ in range(total_steps):
        current_z = sequence[-1]
        # Define the function f(z) = -λz² + λz
        function = np.poly1d([-lmbd, lmbd, 0.0])
        # Find all inverse values using invert_fun
        inverses = invert_fun(function, current_z)
        # Choose one inverse at random
        next_z = random.choice(inverses)
        sequence.append(next_z)
    
    # Return the remaining elements after dropping the first num_dropped
    return sequence[num_dropped:]
    
```

```python
def plot_julia(julia_points, lmbd):
    """
    Input: List of complex numbers that lie on the Julia set for given lmbd
    
    Output: Scatter plot of complex number in list julia_points
    """
    
    fig, axs = plt.subplots()
    
    # Extract real and imaginary parts of the complex numbers
    real = [z.real for z in julia_points]
    imag = [z.imag for z in julia_points]
    
    # Create scatter plot with small points
    axs.scatter(real, imag, s=1, marker='.', color='black')
    
    # Set title with lambda value using Greek letter
    axs.set_title(f'Julia Set for λ = {lmbd.real:.2f} + {lmbd.imag:.2f}i')
    
    # Label axes
    axs.set_xlabel('Real Part')
    axs.set_ylabel('Imaginary Part')
    
    # Ensure equal aspect ratio for correct visualization
    axs.set_aspect('equal')
    
    return fig
```

```python
##############################################################################
# Part 2 - Compute and plot Mandelbrot sets

MAX_ITERS = 20


def iterate_mandel(z_0):
    """
    Input: complex z_0
    
    Output: number of iterations for Mandelbrot function f(z) = z ** 2 + z_0 to diverge,
    maximum number of iterations is capped at MAX_ITERS
    """
    z_function = 0  # Initial value of z is 0
    for k_output in range(MAX_ITERS):
        z_function = z_function ** 2 + z_0
        if abs(z_function) > 2:
            return k_output
    return MAX_ITERS
    
```

```python
def mandel_table(real_values, imag_values):
    """
    Input: arrays real_values, imag_values of floats
    
    Output: 2D numpy array indicating number of iterations for
    divergence of Mandelbrot iteration initialized at
    corresponding complex values
    """
    real_grid, imag_grid = np.meshgrid(real_values, imag_values)
    c_values = real_grid + 1j * imag_grid
    
    # Vectorize the iterate_mandel function to apply to all elements
    vfunc = np.vectorize(iterate_mandel)
    iterations = vfunc(c_values)
    
    return iterations
```

```python
def plot_mandelbrot(real_values, imag_values):
    """
    Input: lists (or numpy arrays) real_values and imag_values
    
    Output: matplotlib figure of image plot of mandel_table()
    applied to real_values, imag_values
    """
    
    fig, axs = plt.subplots()
    
    mandel_data = mandel_table(real_values, imag_values)
    
    real_min, real_max = np.min(real_values), np.max(real_values)
    imag_min, imag_max = np.min(imag_values), np.max(imag_values)
    
    # Create the image with correct orientation and aspect ratio
    image = axs.imshow(mandel_data,
                       origin='lower',
                       extent=[real_min, real_max, imag_min, imag_max],
                       aspect='equal',
                       cmap='hot')  # Using a colormap for better visualization
    
    axs.set_title("Mandelbrot Set")
    axs.set_xlabel("Real Part")
    axs.set_ylabel("Imaginary Part")
    
    # Add colorbar to show iteration scale
    plt.colorbar(image, ax=axs, label='Iterations')
    
    return fig
```

```python
#########################################################################################
# Part 3 - Compute and plot Newton basins 

EPSILON = 0.000001

def newton_index(roots, z_0):
    """
    Input: List of complex roots, complex initial value z_0
    
    Output: Index of root in roots to which Newton iteration with initial
    value z_0 converges (with tolerance EPSILON)
    
    NOTE: If the iteration encounters a zero derivative, return len(roots)
    """
    
    # Handle edge case of empty roots list
    if not roots:
        return 0
    
    # Construct the polynomial from the roots and compute its derivative
    coeffs = np.poly(roots)
    g_function = np.poly1d(coeffs, variable='z')
    g_prime = g_function.deriv()
    
    z_current = z_0
    
    while True:
        # Evaluate the current polynomial and its derivative
        g_z = g_function(z_current)
        gp_z = g_prime(z_current)
        
        # Check termination conditions
        if abs(gp_z) <= EPSILON:
            return len(roots)
        if abs(g_z) <= EPSILON:
            # Find the closest root(s)
            distances = [abs(z_current - root) for root in roots]
            min_distance = min(distances)
            # Get indices of all roots with the minimum distance
            candidates = [i for i, d in enumerate(distances) if d == min_distance]
            return min(candidates)
        
        # Update the current guess using Newton-Raphson iteration
        z_current -= g_z / gp_z
```

```python
def newton_table(roots, real_values, imag_values):
    """
    Input: List of complex numbers roots, two lists of floats that define a grid
    of complex numbers
    
    Output:  2D numpy array of Newton indices where z_0 takes on values in
    the grid of complex numbers
    """
    
    # Create grid of complex numbers from real and imaginary values
    real_grid, imag_grid = np.meshgrid(real_values, imag_values)
    complex_grid = real_grid + 1j * imag_grid
    
    # Vectorize the newton_index function to apply over the grid
    vectorized_newton = np.vectorize(lambda z: newton_index(roots, z))
    
    # Apply the function to each element in the grid
    indices = vectorized_newton(complex_grid)
    
    return indices
```

```python
#########################################################################################
# Student code for plotting Newton basins (peer-graded)

def plot_newton(roots, real_values, imag_values):
    """
    Input: List of complex numbers roots, two lists of floats that define a grid
    of complex numbers
    
    Output: matplotlib figure of mage plot of mandel_table()
    applied to roots, real_values, imag_values
    """
    
    # Compute the Newton indices for the grid
    newton_data = newton_table(roots, real_values, imag_values)
    
    # Create figure and axes
    fig, axs = plt.subplots()
    
    # Determine the plot extent
    real_min, real_max = np.min(real_values), np.max(real_values)
    imag_min, imag_max = np.min(imag_values), np.max(imag_values)
    
    # Plot the Newton basins using imshow
    image = axs.imshow(newton_data,
                       origin='lower',
                       extent=[real_min, real_max, imag_min, imag_max],
                       aspect='equal',
                       cmap='nipy_spectral')  # Using a colormap for better distinction
    
    # Overlay the roots as red dots
    real_parts = [root.real for root in roots]
    imag_parts = [root.imag for root in roots]
    axs.scatter(real_parts, imag_parts, color='red', s=40, marker='o', edgecolors='black')
    
    # Set title and labels
    axs.set_title(f'Newton Basins for {len(roots)} Roots')
    axs.set_xlabel('Real Part')
    axs.set_ylabel('Imaginary Part')
    
    # Add a colorbar to indicate the index values
    plt.colorbar(image, ax=axs, label='Root Index')
    
    return fig
```
