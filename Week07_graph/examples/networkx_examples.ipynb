```python
"""
Draw some simple graphs using various layout options in networkx
"""

import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Check the version of the installed networkx package 
print(nx.__version__)
```

```python
# Create a simple graph
grph = nx.Graph()
grph.add_nodes_from([2, 3, 5, 7])
grph.add_edges_from ([(2, 3), (3, 5), (3, 7), (5, 7)])
print(grph.nodes)
print(grph.edges)
```

```python
# Draw a simple graph with specified layout
node_pos = {2 : (0, 0), 3 : (1, 0), 5 : (2, 0), 7 : (2, 1)}
nx.draw(grph, pos=node_pos, with_labels=True, node_color="y")
```

```python
# Draw dodecahedron using spring layout
dodecahedral = nx.dodecahedral_graph()
nx.draw(dodecahedral, with_labels=True, node_color="y")
```
